```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the dataset
student_data = pd.read_csv('student-alcohol-consumption.csv', index_col=0)
```

```python
student_data
```

```python
student_data.isna()
```

```python
# highest study time
study_time = student_data.groupby("school")[["study_time"]].max()
school = student_data.sort_values(by="study_time", ascending=False)
school.head()
```

```python
# a scatter plot of absences vs. final grade
sns.scatterplot(x="absences", y="G3",
                data=student_data,
                hue="location")
plt.show()
```

```python
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location",
                hue_order=["Rural", "Urban"])
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}
```

```python
# a count plot of school with location subgroups
sns.countplot(x="school", data=student_data,
              hue="location",
              palette=palette_colors)
plt.show()
```

```python
# use relplot() instaed of scatterplot()
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter")
plt.show()
```

```python
# make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter",
            col="study_time")
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")
plt.show()
```

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter")
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"])
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])
plt.show()
```
