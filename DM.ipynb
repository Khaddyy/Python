```python
# Identify outliers
Q1 = df['Glucose'].quantile(0.25)
Q3 = df['Glucose'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Glucose'] < lower_bound) | (df['Glucose'] > upper_bound)]
print("Outliers:")
print(outliers)
```
