```python
column_names = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# Calculate the sum of specified columns
column_sums = covid[column_names].sum().plot(kind="bar")
```

```python
#sum of region
Region=covid.groupby(by='WHO Region').sum()
Region
```

```python
#plot a bar of confirmed cases
Region['Confirmed'].plot(kind='bar', ) 
plt.title(' Confirmed Cases')
```

```python
#plot a bar of deaths recorded
Region['Deaths'].plot(kind='bar', ) 
plt.title(' Deaths Recorded')
```

```python
#plot a chart showing the no of recoveries
Region['Recovered'].plot(kind='pie',  figsize=(10,5), shadow=True) 
plt.title(' Recovered')
```

```python
#plot a chart showing the % of recoveries
Region['Recovered'].plot(kind='pie',  figsize=(10,5), shadow=True, autopct='%1.1f%%') 
plt.title(' Recovered')
```

```python
#plot a chart showing the % of active cases
Region['Active'].plot(kind='pie',  figsize=(10,5), shadow=True, autopct='%1.1f%%') 
plt.title(' Active')
```
